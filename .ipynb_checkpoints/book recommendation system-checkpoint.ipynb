```python
books['Book-Author'] = books['Book-Author'].fillna('Anonymous')
books['Publisher'] = books['Publisher'].fillna('Anonymous')
```

```python
books.isna().sum()
```

### popularity recommendation system

```python
ratings_with_titles = ratings.merge(books,on='ISBN')
```

```python
num_ratings = ratings_with_titles.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings = num_ratings.rename(columns={'Book-Rating':'num_of_ratings'})
num_ratings
```

```python
avg_ratings = ratings_with_titles.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings = avg_ratings.rename(columns={'Book-Rating':'avg_ratings'})
avg_ratings
```

```python
popular_books = num_ratings.merge(avg_ratings,on='Book-Title')
popular_books
```

```python
popular_books = popular_books[popular_books['num_of_ratings'] >= 250].sort_values('avg_ratings',ascending=False).head(50)
popular_books
```

```python
popular_50 = popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')
popular_50
```

```python
popular_50.columns
```

```python
popular_50 = popular_50[['Book-Title', 'num_of_ratings', 'avg_ratings', 'Book-Author',
       'Image-URL-M']]
popular_50
```

```python
popular_50.shape
```

### Collaborative filtering based recommendation system

```python
x = ratings_with_titles.groupby('User-ID').count()['Book-Rating'] > 200
users_gave_more_ratings = x[x].index
```

```python
filtered_rating = ratings_with_titles[ratings_with_titles['User-ID'].isin(users_gave_more_ratings)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.shape
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```

```python
recommend('You Belong To Me')
```
